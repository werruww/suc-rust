```python
# هذا الأمر يقوم بإنشاء أو الكتابة فوق ملف src/main.rs بالمحتوى التالي
# تأكد من وجود مجلد src أولاً: mkdir -p src
cat << 'EOF' > src/main.rs
use std::env;
use std::io::{self, Write};

use dotenv::dotenv;
use reqwest::Client;
use serde::{Deserialize, Serialize};

// هيكل الطلب
#[derive(Serialize)]
struct HFRequest {
    inputs: String,
    // parameters: Option<GenerationParameters>, // يمكن إضافتها لاحقًا
}

// هيكل الرد المتوقع
#[derive(Deserialize, Debug)]
struct HFResponseItem {
    generated_text: String,
}

#[tokio::main]
async fn main() -> Result<(), Box<dyn std::error::Error>> {
    dotenv().ok();
    let hf_token = env::var("HF_TOKEN")
        .expect("لم يتم العثور على HF_TOKEN في البيئة أو ملف .env");

    let client = Client::new();
    // استخدام نموذج أصغر ومناسب للطبقة المجانية
    let model_id = "Qwen/Qwen2-7B-Instruct";
    let model_url = format!("https://api-inference.huggingface.co/models/{}", model_id);
    println!("ℹ️ استخدام النموذج: {}", model_id);
    println!("🔗 عبر الرابط: {}", model_url);

    println!("🤖 مرحبًا بك في شات ذكاء صناعي بلغة Rust (أكتب 'exit' للخروج)");

    loop {
        print!("🧑‍💻 أنت: ");
        io::stdout().flush()?;

        let mut input = String::new();
        io::stdin().read_line(&mut input)?;
        let input = input.trim();

        if input.is_empty() {
            continue;
        }
        if input.eq_ignore_ascii_case("exit") {
            println!("👋 إلى اللقاء!");
            break;
        }

        let request_payload = HFRequest {
            inputs: input.to_string(),
        };

        println!("⏳ جارٍ إرسال الطلب إلى النموذج...");

        match client
            .post(&model_url)
            .bearer_auth(&hf_token)
            .json(&request_payload)
            .send()
            .await
        {
            Ok(resp) => {
                // احصل على الحالة أولاً قبل استهلاك resp
                let status = resp.status();

                if status.is_success() {
                    // محاولة قراءة JSON (تستهلك resp)
                    match resp.json::<Vec<HFResponseItem>>().await {
                        Ok(json_resp) => {
                            if let Some(item) = json_resp.get(0) {
                                let generated = item.generated_text.trim();
                                let output = generated.strip_prefix(input).unwrap_or(generated).trim();
                                println!("🤖 النموذج: {}", output);
                            } else {
                                println!("⚠️ الرد كان مصفوفة فارغة.");
                            }
                        }
                        Err(e) => {
                            println!("⚠️ فشل في فك JSON، خطأ: {}", e);
                            println!("ℹ️ لا يمكن عرض النص الخام لأن محاولة فك JSON فشلت بعد استهلاك الرد.");
                        }
                    }
                } else {
                    // الحالة لم تكن ناجحة، اقرأ النص كخطأ (يستهلك resp)
                    println!("❌ فشل الطلب، حالة HTTP: {}", status); // استخدم الحالة المحفوظة
                    let error_text = resp.text().await.unwrap_or_else(|e| {
                        format!("فشل في قراءة نص الخطأ أيضًا: {}", e)
                    });
                    println!("📄 رسالة الخطأ من الخادم:\n{}", error_text);

                    // استخدم الحالة المحفوظة للتحقق
                    let status_code = status.as_u16();
                    if status_code == 503 {
                        println!("ℹ️ خطأ 503 يعني أن النموذج قد يكون قيد التحميل (جرب مرة أخرى بعد دقيقة) أو غير متاح حاليًا.");
                    } else if status_code == 401 {
                        println!("ℹ️ خطأ 401 يعني مشكلة في التوثيق. تأكد من صحة HF_TOKEN.");
                    } else if status_code == 400 {
                         println!("ℹ️ خطأ 400 يعني طلبًا غير صالح (Bad Request). قد تكون البارامترات أو صيغة الإدخال غير صحيحة للنموذج.");
                    } else if status_code == 422 {
                        println!("ℹ️ خطأ 422 (Unprocessable Entity) يعني غالبًا أن الإدخال غير مفهوم بواسطة النموذج.");
                    }
                    // يمكنك إضافة المزيد من حالات الخطأ هنا
                }
            }
            Err(e) => {
                // خطأ في الاتصال نفسه
                println!("❌ حدث خطأ أثناء إرسال الطلب: {}", e);
            }
        }
        println!("------------------------------------");
    }

    Ok(())
}
EOF
```

```python
!mkdir -p /content/rust_ai_chat/src
!cat << 'EOF' > /content/rust_ai_chat/src/main.rs
use std::env;
use std::io::{self, Write};

use dotenv::dotenv;
use reqwest::Client;
use serde::{Deserialize, Serialize};

// Request structure
#[derive(Serialize)]
struct HFRequest {
    inputs: String,
    // parameters: Option<GenerationParameters>, // Can be added later
}

// Expected response structure
#[derive(Deserialize, Debug)]
struct HFResponseItem {
    generated_text: String,
}

#[tokio::main]
async fn main() -> Result<(), Box<dyn std::error::Error>> {
    dotenv().ok();
    let hf_token = env::var("HF_TOKEN")
        .expect("HF_TOKEN not found in environment or .env file");

    let client = Client::new();
    // Use a smaller model suitable for the free tier
    let model_id = "Qwen/Qwen2-7B-Instruct";
    let model_url = format!("https://api-inference.huggingface.co/models/{}", model_id);
    println!("ℹ️ Using model: {}", model_id);
    println!("🔗 Via URL: {}", model_url);

    println!("🤖 Welcome to the Rust AI Chat (type 'exit' to quit)");

    loop {
        print!("🧑‍💻 You: ");
        io::stdout().flush()?;

        let mut input = String::new();
        io::stdin().read_line(&mut input)?;
        let input = input.trim();

        if input.is_empty() {
            continue;
        }
        if input.eq_ignore_ascii_case("exit") {
            println!("👋 Goodbye!");
            break;
        }

        let request_payload = HFRequest {
            inputs: input.to_string(),
        };

        println!("⏳ Sending request to the model...");

        match client
            .post(&model_url)
            .bearer_auth(&hf_token)
            .json(&request_payload)
            .send()
            .await
        {
            Ok(resp) => {
                // Get status first before consuming resp
                let status = resp.status();

                if status.is_success() {
                    // Try to read JSON (consumes resp)
                    match resp.json::<Vec<HFResponseItem>>().await {
                        Ok(json_resp) => {
                            if let Some(item) = json_resp.get(0) {
                                let generated = item.generated_text.trim();
                                let output = generated.strip_prefix(input).unwrap_or(generated).trim();
                                println!("🤖 Model: {}", output);
                            } else {
                                println!("⚠️ Response was an empty array.");
                            }
                        }
                        Err(e) => {
                            println!("⚠️ Failed to decode JSON, error: {}", e);
                            println!("ℹ️ Cannot display raw text because JSON decoding attempt failed after consuming the response.");
                        }
                    }
                } else {
                    // Status was not successful, read text as error (consumes resp)
                    println!("❌ Request failed, HTTP status: {}", status); // Use saved status
                    let error_text = resp.text().await.unwrap_or_else(|e| {
                        format!("Failed to read error text as well: {}", e)
                    });
                    println!("📄 Server error message:\n{}", error_text);

                    // Use saved status for checks
                    let status_code = status.as_u16();
                    if status_code == 503 {
                        println!("ℹ️ Error 503 means the model might be loading (try again in a minute) or is currently unavailable.");
                    } else if status_code == 401 {
                        println!("ℹ️ Error 401 means an authentication issue. Check your HF_TOKEN.");
                    } else if status_code == 400 {
                         println!("ℹ️ Error 400 means Bad Request. Parameters or input format might be incorrect for the model.");
                    } else if status_code == 422 {
                        println!("ℹ️ Error 422 (Unprocessable Entity) often means the input was not understood by the model.");
                    }
                    // Add more error cases here if needed
                }
            }
            Err(e) => {
                // Error with the connection itself
                println!("❌ An error occurred while sending the request: {}", e);
            }
        }
        println!("------------------------------------");
    }

    Ok(())
}
EOF
```

```python
rust
use std::error::Error;

use ollama_rs::{generation::completion::GenerationRequest, Ollama};

fn main() -> Result<(), Box<dyn Error>> {
    let model = "llama2:latest".to_string();
    let prompt = "Why is the sky blue?".to_string();

    let ollama = Ollama::new()?;

    let res = ollama.generate(GenerationRequest::new(model, prompt)).unwrap();

    if let Ok(res) = res {
        println!("{}", res.response);
    } else {
        eprintln!("Error: {:?}", res.err());
    }

    Ok(())
}
```

```python
//! Completion streaming using Callback API
//!
//! In this example, you should see a streamed text output from your local LLM,
//! just like you saw on ChatGPT =)

use std::io::Write;

use ollama_rest::{models::generate::{GenerationRequest, GenerationResponse}, Ollama};
use serde_json::json;

#[tokio::main]
async fn main() {
    // Make sure Ollama serves at 127.0.0.1:11434
    let ollama = Ollama::default();

    let request = serde_json::from_value::<GenerationRequest>(json!({
        "model": "llama3.2:1b",
        "prompt": "Why is the sky blue?",
    })).unwrap();

    let final_res = ollama.generate(
        &request,
        Some(|res: &GenerationResponse| {
            if !res.done {
                print!("{}", res.response);
                // Flush stdout for each word to allow realtime output
                std::io::stdout().flush().unwrap();
            }
        })
    ).await.unwrap();

    println!("\n\nFinal response:");
    println!("{final_res:?}");
}
```

```python

use std::io::Write;
use futures::stream::StreamExt; // Import StreamExt
use ollama_rest::{
    models::generate::GenerationRequest,
    Ollama
};
use serde_json::json;

#[tokio::main]
async fn main() -> anyhow::Result<()> {
    let ollama = Ollama::default();

    let request = serde_json::from_value::<GenerationRequest>(json!({
        "model": "llama2:latest",
        "prompt": "Why is the sky blue?",
    }))
    .unwrap();

    let mut stream = ollama.generate_streamed(&request).await?;

    while let Some(result) = stream.try_next().await? {
        // Use try_next and handle Result
        if !result.done {
            print!("{}", result.response);
            std::io::stdout().flush()?;
        }
    }

    println!();
    Ok(())
}
```
